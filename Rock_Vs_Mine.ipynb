IMPORTING REQUIREMENTS


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND PROCESSING THE DATA


```python
#LOADING THE DATASET TO PANDAS DATAFRAME
sonar_data=pd.read_csv('/content/sonar data.csv', header=None)
```

```python
#CHECKING FIRST 5 LINES
sonar_data.head()
```

```python
#number of rows and columns
sonar_data.shape
```

```python
#statistical meausres of the data
sonar_data.describe()
```

```python
#how many rocks and mines are there in the data
sonar_data[60].value_counts()
```

```python
#grouping data
sonar_data.groupby(60).mean()
```

```python
#seperating data and label
X=sonar_data.drop(columns=60,axis=1)
Y=sonar_data[60]
```

TRAINING AND TESTING DATA


```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
#checking the shape of train and test datas
print(X.shape,Xtrain.shape,Xtest.shape)
```

MODEL TRAINING--> Logistic Regression model is used here , cause it gives better results for binary values


```python
model=LogisticRegression()
```

```python
#training LogisticRegression with training data
model.fit(Xtrain,Ytrain)
```

MODEL EVALUATION...


```python
#accuracy on test data
Xtrain_prediction=model.predict(Xtrain)
training_data_accuracy=accuracy_score(Xtrain_prediction,Ytrain)
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
#accuracy on test data
Xtest_prediction=model.predict(Xtest)
test_data_accuracy=accuracy_score(Xtest_prediction,Ytest)
```

```python
print('Accuracy on test data',test_data_accuracy)
```
